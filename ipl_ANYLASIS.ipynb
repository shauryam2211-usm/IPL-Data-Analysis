```python
# using isin fn 
# find the most destructive death over batsman in thr history of ipl
# strike rate= (number of runs / number of ball)*100
# min batsman 200 balls in over 16-20
```

```python
death_over=delivery[delivery['over']>15]
```

```python
all_batsman = death_over.groupby('batsman')['batsman_runs'].count()
x = all_batsman>200
batsman_list=all_batsman[x].index.tolist()
```

```python
# runs scored by all 43 batsman 
#balls played by these 43 batsman
```

```python
final =death_over[death_over['batsman'].isin(batsman_list)]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls = final.groupby('batsman').size()
```

```python
sr=(runs/balls)*100
```

```python
sr.sort_values(ascending=False).head(10)
```
